```python
import requests
import json
import pandas as pd
import http.client
import sys
import os
sys.path.append(os.path.abspath('../'))
import dotenv 
dotenv.load_dotenv()
key = os.getenv("key")
```

```python
def get_peliculas(genre, titleType, page, key):

    resultados = []
    try:
        connection = http.client.HTTPSConnection("moviesdatabase.p.rapidapi.com")

        headers = {
            'x-rapidapi-key': key,
            'x-rapidapi-host': "moviesdatabase.p.rapidapi.com"
        }

        url_data = f"/titles?genre={genre}&startYear=1990&titleType={titleType}&sort=year.incr&page={page}"
        
        connection.request("GET", url_data, headers=headers)
        res = connection.getresponse()
        
        if res.status == 200:
            data = res.read()
            dicc_datos = json.loads(data.decode("utf-8"))
            results = dicc_datos.get('results', [])

            for r in results:
                movie = {
                    "id": r.get('id', "N/A"),
                    "titleText": r.get('titleText', {}).get('text', "N/A"),
                    "releaseYear": r.get('releaseYear', {}).get('year', "N/A")
                }
                resultados.append(movie)
        else:
            print(f"Error en la solicitud: {res.status} {res.reason}")

    except Exception as e:
        print(f"Error al hacer la petición a la API: {e}")

    return pd.DataFrame(resultados)
```

```python
df_drama = get_peliculas('Drama','movie', 1, key)
df_comedia = get_peliculas('Comedy','movie', 1, key)
df_accion = get_peliculas('Action','movie', 1, key)
df_Fantasy = get_peliculas('Fantasy','movie', 1, key)
df_Horror = get_peliculas('Horror','movie', 1, key)
df_Mystery = get_peliculas('Mystery','movie', 1, key)
df_Romance = get_peliculas('Romance','movie', 1, key)
df_Thriller = get_peliculas('Thriller','movie', 1, key)
df_drama1 = get_peliculas('Drama','movie', 2, key)
df_comedia1 = get_peliculas('Comedy','movie', 2, key)
df_accion1 = get_peliculas('Action','movie', 2, key)
df_Fantasy1 = get_peliculas('Fantasy','movie', 2, key)
df_Horror1 = get_peliculas('Horror','movie', 2, key)
df_Mystery1 = get_peliculas('Mystery','movie', 2, key)
df_Romance1 = get_peliculas('Romance','movie', 2, key)
df_Thriller1 = get_peliculas('Thriller','movie', 2, key)

dataframes = {
    "Drama": df_drama,
    "Comedy": df_comedia,
    "Action": df_accion,
    "Fantasy": df_Fantasy,
    "Horror": df_Horror,
    "Mystery": df_Mystery,
    "Romance": df_Romance,
    "Thriller": df_Thriller,
    "Drama": df_drama,
    "Comedy": df_comedia,
    "Action": df_accion,
    "Fantasy": df_Fantasy,
    "Horror": df_Horror,
    "Mystery": df_Mystery,
    "Romance": df_Romance,
    "Thriller": df_Thriller}

for name, df in dataframes.items():
    df["Categoria"] = name
```
